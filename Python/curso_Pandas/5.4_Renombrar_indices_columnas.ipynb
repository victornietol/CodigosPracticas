```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis', 'Maria'],
    'Salario': [50000, 60000, 55000, 58000, 62000],
    'Años_Experiencia': [5, 7, 6, 8, 9]
}

df = pd.DataFrame(data)
df
```

```python
df_renombrado = df.rename(columns={'Salario': 'Salario_Anual', 'Años_Experiencia': 'Experiencia'})
df_renombrado
```

```python
df.columns = ['Empleado', 'Salario_Anual', 'Experiencia']
df
```

```python
df_renombrado_indices = df.rename(index={0: 'Emp1', 1: 'Emp2', 2: 'Emp3', 3: 'Emp4', 4: 'Emp5'})
df_renombrado_indices
```

```python
df.index = ['Emp1', 'Emp2', 'Emp3', 'Emp4', 'Emp5']
df
```

```python
df_reorganizado = df.reindex(columns=['Experiencia', 'Salario_Anual', 'Empleado'])
df_reorganizado
```

```python
df = df[['Experiencia', 'Salario_Anual', 'Empleado']]
df
```

```python
df_reorganizado_filas = df.reindex(index=['Emp3', 'Emp1', 'Emp5', 'Emp4', 'Emp2'])
df_reorganizado_filas
```
