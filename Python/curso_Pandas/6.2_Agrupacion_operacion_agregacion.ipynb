```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan', 'Ana', 'Pedro', 'Luis', 'Maria', 'Ana', 'Pedro', 'Luis', 'Maria'],
    'Departamento': ['Ventas', 'Marketing', 'Ventas', 'IT', 'Marketing', 'IT', 'Ventas', 'Marketing', 'Ventas'],
    'Salario': [50000, 60000, 55000, 58000, 62000, 60000, 55000, 58000, 62000],
    'Años_Experiencia': [5, 7, 6, 8, 9, 7, 6, 8, 9]
}
df = pd.DataFrame(data)
display(df)
```

```python
suma_salarios = df.groupby('Departamento')['Salario'].sum()
display(suma_salarios)
```

```python
media_experiencia = df.groupby('Departamento')['Años_Experiencia'].mean()
display(media_experiencia)
```

```python
agregaciones_salario = df.groupby('Departamento')['Salario'].agg(['sum', 'mean', 'max'])
display(agregaciones_salario)
```

```python
agregaciones_multiples = df.groupby('Departamento').agg({
    'Salario': ['sum', 'mean', 'max'],
    'Años_Experiencia': ['mean', 'min', 'max']
})
display(agregaciones_multiples)
```

```python
agregaciones_complejas = df.groupby([ 'Departamento','Empleado']).agg({
    'Salario': ['sum', 'mean'],
    'Años_Experiencia': 'mean'
})
display(agregaciones_complejas)
```
