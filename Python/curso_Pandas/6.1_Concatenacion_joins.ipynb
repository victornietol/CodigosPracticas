```python
import pandas as pd
```

```python
data1 = {
    'Empleado': ['Juan', 'Ana'],
    'Salario': [50000, 60000]
}

data2 = {
    'Empleado': ['Pedro', 'Luis'],
    'Salario': [55000, 58000]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

df_concatenado_filas = pd.concat([df1, df2], axis=0)

display(df1)
display(df2)
display(df_concatenado_filas)
```

```python
data3 = {
    'Empleado': ['Juan', 'Ana'],
    'Departamento': ['Ventas', 'Marketing']
}

data4 = {
    'Empleado': ['Juan', 'Ana'],
    'Salario': [50000, 60000]
}

df3 = pd.DataFrame(data3)
df4 = pd.DataFrame(data4)

df_concatenado_columnas = pd.concat([df3, df4], axis=1)

display(df3)
display(df4)
display(df_concatenado_columnas)
```

```python
data_left = {
    'Empleado': ['Juan', 'Ana', 'Pedro'],
    'Salario': [50000, 60000, 55000]
}

data_right = {
    'Empleado': ['Ana', 'Pedro', 'Luis'],
    'Departamento': ['Marketing', 'Ventas', 'IT']
}

df_left = pd.DataFrame(data_left)
df_right = pd.DataFrame(data_right)

df_inner = pd.merge(df_left, df_right, on='Empleado', how='inner')

display(df_left)
display(df_right)
display(df_inner)
```

```python
df_left_join = pd.merge(df_left, df_right, on='Empleado', how='left')
display(df_left_join)
```

```python
df_right_join = pd.merge(df_left, df_right, on='Empleado', how='right')
display(df_right_join)
```

```python
df_outer_join = pd.merge(df_left, df_right, on='Empleado', how='outer')
display(df_outer_join)
```
