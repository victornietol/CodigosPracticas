# Anagrama

```python
def anagrama(word: str) -> bool:
    length = len(word)
    for i in range((length//2)+1): # hasta la mitad
        if (word[i] != word[length-i-1]):
            return False
    return True

word = "aannaa"
print(anagrama(word))

word = "aansnaa"
print(anagrama(word))

word = "aanscnaa"
print(anagrama(word))
```
