# Anagrama

```python
def anagrama(word: str) -> bool:
    length = len(word)
    for i in range((length//2)+1): # hasta la mitad
        if (word[i] != word[length-i-1]):
            return False
    return True

word = "aannaa"
print(anagrama(word))

word = "aansnaa"
print(anagrama(word))

word = "aanscnaa"
print(anagrama(word))
```

# Factorial

```python
def fact(n: int) -> int:
    count = 1
    for i in range(1, n+1):
        count *= i
    return count

def fact_rec(n: int) -> int: # recursivo
    if n==1:
        return 1
    return n * fact_rec(n-1)

n = 8
print(fact(n))
print(fact_rec(n))
```

# Fibonacci 0,1,1,2,3,5,8,...

```python
def fib(n: int) -> int: # un elemento de la serie (indices del 1 en adelante)
    elements = [0,1]
    n -= 1
    if n<=1 and n>-1:
        return elements[n]
    
    for _ in range(n-1):
        res = elements[0] + elements[1]
        elements[0] = elements[1]
        elements[1] = res

    return elements[1]

def fib_rec(n: int): # un elemento de la serie (indices del 0 en adelante)
    if n==0 or n==1:
        return n
    
    return fib_rec(n-2) + fib_rec(n-1)


def serie_fib(n_elem: int) -> list: # regresa la serie con el numero de elementos indicados
    elements = [0,1]
    if n_elem == 1:
        return [0]
    if n_elem == 2:
        return elements
    
    for i in range(1, n_elem-1):
        elements.append(elements[i-1] + elements[i])
    
    return elements

n = 7
print(fib(n))
print(fib_rec(n-1))
print(serie_fib(7))
```

# Baseball Game

- You are keeping the scores for a baseball game with strange rules. At the beginning of the game, you start with an empty record.

- You are given a list of strings operations, where operations[i] is the ith operation you must apply to the record and is one of the following:

- An integer x: Record a new score of x.
- '+': Record a new score that is the sum of the previous two scores.
- 'D': Record a new score that is the double of the previous score.
- 'C': Invalidate the previous score, removing it from the record.
- Return the sum of all the scores on the record after applying all the operations.

- The test cases are generated such that the answer and all intermediate calculations fit in a 32-bit integer and that all operations are valid.

```python
def baseball_game(ops: list) -> int:
    record = []
    for score in ops:
        if score == "+":
            record.append(record[-2] + record[-1])
        elif score =="D":
            record.append(2*(record[-1]))
        elif score == "C":
            record.pop()
        else:
            record.append(int(score))
    return sum(record)

ops = ["5","2","C","D","+"]
res = baseball_game(ops)
print(ops, "\noutput = ", res, end="\n\n")

ops = ["5","-2","4","C","D","9","+","+"]
res = baseball_game(ops)
print(ops, "\noutput = ", res, end="\n\n")

ops = ["1","C"]
res = baseball_game(ops)
print(ops, "\noutput = ", res)
```

# Lista de tareas (básico)

```python
class ToDoList:
    def __init__(self):
        self.__id = 1
        self.__tasks = {} ### tareas en un diccionario con su respectivo id para agilizar busqueda

    def __sum_id(self):
        self.__id += 1

    def __create_taskn_name(self) -> str:
        return input("Nombre de la tarea: ")

    def __create_task_description(self) -> str:
        return input("Descripcion de la tarea: ")
    
    def list_tasks(self):
        if not self.__tasks:
            print("No se han creado tareas.\n")
        else:
            print("Lista de tareas:")
            for key, value in self.__tasks.items():
                print(f"{value}")

    def create_new_task(self):
        try:
            name = self.__create_taskn_name()
            description = self.__create_task_description()
            task = Task(self.__id, name, description)
            self.__tasks[task.id] = task
        except Exception as e:
            print("Error al crear nueva tarea.")
        else:
            self.__sum_id()

    def mark_task_as_completed(self, id):
        try:
            task = self.__tasks[id]
            task.completed = True
        except Exception as e:
            print("Tarea no encontrada.")
        else:
            print(f"Tarea con id={task.id} marcada como COMPLETADA.")

    def eliminate_task(self, id):
        try:
            del self.__tasks[id]
        except Exception as e:
            print(f"Error al eliminar tarea con id={id}")
        else:
            print(f"Tarea con id={id} eliminado.")

    


class Task:
    def __init__(self, id: int, name: str, description: str):
        self.__id = id
        self.__name = name
        self.__description = description
        self.__completed = False

    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name):
        self.__name = name

    @property
    def description(self):
        return self.__description
    
    @description.setter
    def description(self, description):
        self.__description = description

    @property
    def completed(self):
        return self.__completed
    
    @completed.setter
    def completed(self, completed):
        self.__completed = completed

    @property
    def id(self):
        return self.__id
    
    def __str__(self):
        return f"""{{
    "id": {self.__id},
    "name": {self.__name},
    "description": {self.__description},
    "completed": {self.__completed}
}}
    """



def run():
    option = 0
    toDoList = ToDoList()
    while (option != 5):
        option = int(input("""Ingresa una de las siguientes opciones:
                                1 -> Agregar tarea.
                                2 -> Listar tareas existentes.
                                3 -> Marcar como completada una tarea.
                                4 -> Eliminar una tarea.
                                5 -> Salir.
        Opcion: """))
        
        if option == 1:
            toDoList.create_new_task()
        elif option == 2:
            toDoList.list_tasks()
        elif option == 3:
            id = int(input("Ingresa el id de la tarea a marcar como completada: "))
            toDoList.mark_task_as_completed(id)
        elif option == 4:
            id = int(input("Ingresa el id de la tarea a eliminar: "))
            toDoList.eliminate_task(id)
        elif option == 5:
            pass
        else:
            print("Option incorrecta.")


### Pruebas
run()
```

# Ejercicio cajero sencillo

```python
class Cliente:
    def __init__(self, username: str, password: str):
        self.__username = username
        self.__password = password
        self.__saldo = 0.0
        self.__num_transacciones = 0
        self.__transacciones = {}

    @property
    def username(self) -> str:
        return self.__username
    
    @property
    def password(self) -> str:
        return self.__password
    
    @property
    def saldo(self) -> float:
        return self.__saldo
    
    @saldo.setter
    def saldo(self, nuevo_saldo: float):
        self.__saldo = nuevo_saldo

    @property
    def num_transacciones(self) -> int:
        return self.__num_transacciones
    
    @property
    def transacciones(self) -> dict:
        return self.__transacciones
    
    def __sumar_transaccion(self):
        self.__num_transacciones += 1

    def registrar_transaccion(self, descripcion: str):
        self.__sumar_transaccion()
        self.__transacciones[self.__num_transacciones] = descripcion

class Cajero:
    def __init__(self):
        self.__clientes = {}
        self.__cliente_logueado = None

    def __obtener_cliente_logueado(self) -> Cliente:
        return self.__cliente_logueado

    def __obtener_username(self) -> str:
        return input("Ingresa un username: ")
    
    def __obtener_password(self) -> str:
        return input("Ingresa el password: ")
    
    def __registrar_nuevo_cliente(self, cliente: Cliente):
        self.__clientes[cliente.username] = cliente

    def esta_Logueado(self) -> bool:
        return False if self.__cliente_logueado is None else True

    def crear_cliente(self):
        username = self.__obtener_username()
        if username in self.__clientes:
            print("Username invalido.")
        else:
            password = self.__obtener_password()
            cliente = Cliente(username, password)
            self.__registrar_nuevo_cliente(cliente)
            print("Cliente creado correctamente.")

    def iniciar_session(self, username: str, password: str):
        if username in self.__clientes:
            if password == self.__clientes[username].password:
                self.__cliente_logueado = self.__clientes[username]
                print("Inicio de sesion correcto.")
                return
        print("Credenciales invalidas.")

    def cerrar_sesion(self):
        self.__cliente_logueado = None
        print("Sesion cerrada correctamente.")
    
    def consultar_saldo(self) -> float:
        print(f"Saldo disponible: {self.__obtener_cliente_logueado().saldo}")
    
    def depositar(self, cantidad: float):
        if cantidad > 0:
            self.__obtener_cliente_logueado().saldo += cantidad
            self.__obtener_cliente_logueado().registrar_transaccion(f"Se deposito la cantidad de: {cantidad}")
            print(f"Deposito por la cantidad de {cantidad} realizado correctamente.")
        else:
            print("Cantidad invalida.")

    def retirar(self, cantidad: float):
        if cantidad > 0:
            if self.__obtener_cliente_logueado().saldo >= cantidad:
                self.__obtener_cliente_logueado().saldo -= cantidad
                self.__obtener_cliente_logueado().registrar_transaccion(f"Se retiro la cantidad de: {cantidad}")
                print(f"Retiro por la cantidad de {cantidad} realizado correctamente.")
            else:
                print("Saldo insuficiente para realizar un retiro.")
        else:
            print("Cantidad invalida.")
    
    def consultar_transacciones(self):
        for key, value in self.__obtener_cliente_logueado().transacciones.items():
            print("===================================")
            print(f"Transaccion numero: {key}")
            print(f"Descripcion: {value}")


def run_cajero():
    ejecutar = True
    cajero = Cajero()
    while ejecutar:
        opc = int(input("""Elija una opcion:
    1. Iniciar sesion.
    2. Crear cliente.
    3. Salir.
"""))
        if opc == 3:
            ejecutar == False
            break

        elif opc == 2:
            cajero.crear_cliente()

        elif opc == 1:
            username = input("Ingresa el username: ")
            password = input("Ingresa el password: ")
            cajero.iniciar_session(username, password)
            while cajero.esta_Logueado() == True:
                opc1 = int(input("""Elija una opcion:
    1. Consultar saldo.
    2. Depositar saldo.
    3. Retirar saldo.
    4. Consultar transacciones realizadas.
    5. Cerrar sesión.
"""))
                if opc1 == 1:
                    cajero.consultar_saldo()
                elif opc1 == 2:
                    cantidad = float(input("  Ingresa la cantidad a depositar: "))
                    cajero.depositar(cantidad)
                elif opc1 == 3:
                    cantidad = float(input("  Ingresa la cantidad a retirar: "))
                    cajero.retirar(cantidad)
                elif opc1 == 4:
                    cajero.consultar_transacciones()
                elif opc1 == 5:
                    cajero.cerrar_sesion() ## Cambia el valor de cajero.esta_Logueado()
                else:
                    print("Opcion invalida.")
            



### PRUEBAS
run_cajero()
```
